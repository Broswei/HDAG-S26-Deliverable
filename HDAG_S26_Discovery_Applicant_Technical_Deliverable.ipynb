```python
# DO NOT EDIT THIS BLOCK!
# To run a cell, press shift + enter

import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

def generate_graph(s, n, t):
  x, y, coef = datasets.make_regression(n_samples=s, n_features=2, n_informative=1, noise=n, coef=True, random_state=t)
  x = np.interp(x, (x.min(), x.max()), (0, 20))
  y = np.interp(y, (y.min(), y.max()), (20000, 150000))

  plt.ion()
  plt.plot(x,y,'.',label='training data')
  plt.xlabel('Years of experience');plt.ylabel('Salary ($)')
  plt.title('Experience Vs. Salary')

generate_graph(100, 10, 2)
```
